## 2. Data Comprehension and visualisation

|Features|Details|
|---|---|
|Indicator 1|In the first step of the three-way handshake, the time interval between the syn ack packet and the ack packet|
|indicator 2|In the second step of the three-way handshake, the time interval between the syn ack packet and the ack packet|
|Indicator 3|The time interval between the ack packet and the first payload packet in the three-way handshake|
|Indicator 4|The response delay of the first packet with payload after the establishment of TCP for multiple flows in the session|
|Indicator 5|In TCP transmission, the actual delay of transmission from the DPI position to the user terminal|
|Indicator 6|In TCP transmission, the transmission delay from the DPI position to the website|
|Indicator 7|In TCP transmission, the percentage of downlink retransmitted packets in the current session|
|Indicator 8|In TCP transmission, the percentage of upstream retransmission packets of the current session|



The three-pass handshake is a communication procedure used by TCP/IP communication protocols to establish a connection between two hosts. The procedure consists of three messages:

* The SYN (synchronization) message is sent by the host initiating the connection.
* The SYN-ACK (synchronization-acknowledgement) message is sent by the destination host in response to the SYN message.
* The ACK message (acknowledgment) is sent by the host initiating the connection in response to the SYN-ACK message.

Flags 1, 2, and 3 measure the time it takes for SYN, SYN-ACK, and ACK messages to be exchanged between the two hosts.

Indicator 4 measures the time it takes for the first payload packet to be transmitted after the TCP connection is established.

Indicators 5 and 6 measure the transmission delay between the DPI (Deep Packet Inspection) point and the user's terminal or website.

Indicators 7 and 8 measure the percentage of packets retransmitted downstream or upstream of the current TCP session. Retransmitted packets are packets that were lost or corrupted during transmission.

```python
def missing_val(data):
    """
    Determine rate of missing values in the dataset
    """
    
    miss_list = []
    cols = data.columns
    x = 0
    sum_data = data.shape[0]
    
    for col in cols:
        x = data[col].isnull().sum()
        rate = (x/sum_data)
        if rate != 0:
            print('{} col have missing val! Rate {} = {}'.format(col, col, rate))
            miss_list.append(col)
        if rate == 0:
            break


    comparate = []   
    if miss_list == comparate:
        print('We will not dealing with missing values')
    else:
        miss_list.sort()
        print('\n','-'*25)
        print('\n Columns sorted:')
        print('-'*25)
        return miss_list
```

```python
def outlier_z(data):
    """
    We will calculate method_quantile of all columns and print the rate of outliers in the data
    """

    q1 = np.quantile(data, 0.25)
    q3 = np.quantile(data, 0.75)
    iqr = q3 - q1
    val_lower = q1 - iqr*1.5
    val_upper = q3 + iqr*1.5
    i = np.array(data.index[(data < val_lower) | (data > val_upper)])
    val = np.array(data[i])
    if len(val) == 0:
        print('There are no outliers')
    else:
        return i, val
```

```python
def sum_duplicates_with_index_to_column(df, column, ignore_columns):
    """
    Sums the duplicate values in a DataFrame.

    Args:
        df (DataFrame): The DataFrame.
        column (str): The name of the column whose duplicate values we want to sum.
        ignore_columns (list): A list of column names to not touch.

    Returns:
        DataFrame: The DataFrame with duplicate values replaced by their sum.
    """

    # Groups the DataFrame by the specified column.
    df_grouped = df.groupby(column)

    # Applies the `sum()` function to each group.
    df_grouped = df_grouped.agg(sum)

    # Drops the columns we don't want to touch.
    for column in ignore_columns:
        df_grouped.drop(column, axis=1, inplace=True)

    # Resets the index of the DataFrame so that the column is now a column.
    df_grouped.reset_index(inplace=True)

    # Renames the column to `specifictime`.
    df_grouped.rename(columns={column: "specifictime"}, inplace=True)

    return df_grouped
```

```python
def features_rapport(df, list_w, title):
    for i in range(len(list_w)):
        #Statistics operation
        print("Rapport",list_w[i],"\n", "-"*80)
        min_list_w = np.min(df[list_w[i]])
        max_list_w = np.max(df[list_w[i]])
        mean_list_w = np.mean(df[list_w[i]])
        var_list_w = np.var(df[list_w[i]])

        #Print
        print("The min value of {} is {}".format(list_w[i], min_list_w))
        print("The max value of {} is {}".format(list_w[i], max_list_w))
        print("The mean value of {} is {}".format(list_w[i], mean_list_w))
        print("The variance value of {} is {}".format(list_w[i], var_list_w))
        print("\n \n")

    #Plotting
    n_plots = len(list_w)
    fig, axes = plt.subplots(nrows=n_plots, ncols=1, figsize=(12, 4*n_plots))
    fig.suptitle(title)

    for i in range(n_plots):
        
        ax = axes[i]
        ax.plot(df[list_w[i]], label=list_w[i])
        ax.set_xlabel("Index")
        ax.set_ylabel("Value")
        ax.set_title("Feature: {}".format(list_w[i]))
        ax.legend()

    fig.tight_layout()
    plt.show()
```

```python
#Unsupervised learning KMEANS

from sklearn.cluster import KMeans


wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0, n_init=10)
    kmeans.fit(X_kmeans)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method For Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS Values')
plt.grid(True)
plt.show()
```

We will choose 2 clusters

```python
kmeans_model = KMeans(n_clusters=2, init='k-means++', random_state=0, n_init=10)
y_kmeans = kmeans_model.fit_predict(X_kmeans)

m = np.array(X_kmeans)
plt.scatter(m[y_kmeans == 0, 0], m[y_kmeans == 0, 1], s=80, c='red', label='Cluster 1')
plt.scatter(m[y_kmeans == 1, 0], m[y_kmeans == 1, 1], s=80, c='yellow', label='Cluster 2')
#plt.scatter(m[y_kmeans == 2, 0], m[y_kmeans == 2, 1], s=80, c='blue', label='Cluster 3')
#plt.scatter(m[y_kmeans == 3, 0], m[y_kmeans == 3, 1], s=80, c='green', label='Cluster 4')
plt.scatter(kmeans_model.cluster_centers_[:, 0], kmeans_model.cluster_centers_[:, 1], s=100, c='black', label='Centroids')
plt.title('Clusters of use indicators')
plt.xlabel('Behavior of user')
plt.ylabel('Observations')
plt.legend()
plt.grid(True)
plt.show()
```
